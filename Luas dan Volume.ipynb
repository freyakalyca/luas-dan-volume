```python
print("Perhitungan Luas Permukaan dan Volume Bangun Ruang")
print("Pilih salah satu!")
print("1.) Luas Permukaan dan Volume Kubus")
print("2.) Luas Permukaan dan Volume Balok")
print("3.) Luas Permukaan dan Volume Tabung")
print("4.) Luas Permukaan dan Volume Bola")
print("5.) Luas Permukaan dan Volume Prisma Segitiga")
print("6.) Luas Permukaan dan Volume Kerucut")
print("7.) Luas Permukaan dan Volume Limas Segiempat")

p=int(input('input your option from 1 to 7 ='))

if p==1:
    print('1.) Luas Permukaan dan Volume Kubus ')
    s=int(input('sisi ='))
    luas=s**2*6
    volume=s**3
    print('Luas permukaan =',luas)
    print('Volume bangun =',volume)

elif p==2:
    print('2.) Luas Permukaan dan Volume Balok ')
    p=int(input('panjang ='))
    l=int(input('lebar ='))
    t=int(input('tinggi ='))
    luas=2*(p*l+p*t+l*t)
    volume=p*l*t
    print('Luas permukaan =',luas)
    print('Volume bangun =',volume)

elif p==3:
    print('3.) Luas Permukaan dan Volume Tabung')
    pi=22/7
    r=int(input('jari-jari ='))
    t=int(input('tinggi tabung ='))
    luas=(2*pi*r*t)+(2*pi*r**2)
    volume=pi*r**2*t
    print('Luas permukaan =',luas)
    print('Volume bangun =',volume)

elif p==4:
    print('4.) Luas Permukaan dan Volume Bola')
    pi=22/7
    r=int(input('jari-jari ='))
    luas=4*pi*r**2
    volume=4/3*pi**3
    print('Luas permukaan =',luas)
    print('Volume bangun =',volume)

elif p==5:
    print('5.) Luas Permukaan dan Volume Prisma Segitiga')
    r1=int(input('alas segitiga ='))
    r2=int(input('tinggi segitiga ='))
    r3=int(input('garis miring ='))
    T=int(input('tinggi prisma ='))
    luas=(r1*r2)+((r1+r2+r3)*T)
    volume=(1/2*r1*r2)*T
    print('Luas permukaan =',luas)
    print('Volume bangun =',volume)

elif p==6:
    print('6.) Luas Permukaan dan Volume Kerucut')
    pi=22/7
    r=int(input('jari-jari ='))
    s=int(input('sisi ='))
    t=int(input('tinggi ='))
    luas=pi*r*s+pi*r**2
    volume=1/3*pi*r**2*t
    print('Luas permukaan =',luas)
    print('Volume bangun =',volume)

elif p==7:
    print('7.) Luas Permukaan dan Volume Limas Segiempat')
    s=int(input('sisi ='))
    a=int(input('alas ='))
    t=int(input('tinggi ='))
    luas=s**2+(4*(1/2*a*t))
    volume=1/3*s**2*t
    print('Luas permukaan =',luas)
    print('Volume bangun =',volume)

else:
    print('Sorry, we cant calculate it')
```
